## IRIS Dataset analysis

```python
import seaborn as sns
import matplotlib.pyplot as mp
import pandas as pd
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

## Correlation with Heatmap

```python
df.corr()
```

```python
dataframe = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True)
mp.show()
```

# Joint Plot

### Joint plot between sepal length and sepal width

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
mp.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='reg')
mp.show()
```

## Joint plot between Petal length and Petal Width

```python
sns.jointplot(x='petal_length',y='petal_width',data=df,kind='reg')
mp.show()
```

## Joint Plot between Sepal Length and Petal Width

```python
sns.jointplot(x='sepal_length',y='petal_width',data=df,kind='reg')
mp.show()
```

## Pair Plot

```python
sns.pairplot(df)
mp.show()
```

```python
sns.pairplot(df,hue='species')
mp.show()
```

# Dist Plot

#### Plotting the 4 combinations by subplot

```python
fig = plt.figure(figsize=(14,12))

plt.subplot(221)
plt.title('Dist plot sepal length')
sns.distplot(df['sepal_length'])

plt.subplot(222)
plt.title('Dist plot sepal width')
sns.distplot(df['sepal_width'])

plt.subplot(223)
plt.title('Dist plot petal length')
sns.distplot(df['petal_length'])

plt.subplot(224)
plt.title('Dist plot petal width')
sns.distplot(df['petal_width'])

plt.show()
```

```python
sns.distplot(df['sepal_length'],kde=False,bins=10)
mp.show()
```
